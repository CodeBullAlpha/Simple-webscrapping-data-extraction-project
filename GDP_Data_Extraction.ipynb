# Practice Project: GDP Data Extraction and Processing

This project aims to extract data from a website and process its contents to gain insights such as what are the top 10 countries with the largest economies 

```python
import numpy as np
import pandas as pd

#ignoring warnings
def warn(*args, **kwargs):
    pass
import warnings
warnings.warn = warn
warnings.filterwarnings('ignore')

#Extract information from tables on website
URL="https://web.archive.org/web/20230902185326/https://en.wikipedia.org/wiki/List_of_countries_by_GDP_%28nominal%29"
tables=pd.read_html(URL)
tables
```

```python
#convert the third table to a pageframe
df=pd.DataFrame(tables[3])
# Replace the column headers with column numbers
df.columns = range(df.shape[1])
# Retain columns with index 0 and 2 (name of country and value of GDP quoted by IMF)
df2=df[[0,2]]
# Retain the Rows with index 1 to 10, indicating the top 10 economies of the world.
df3=df2.iloc[1:11,:]
# Assign column names as "Country" and "GDP (Million USD)"
df3.columns=['Country','GDP (Million USD)']
df3
```

```python
# Change the data type of the 'GDP (Million USD)'
df3['GDP (Million USD)']=df3['GDP (Million USD)'].astype(int)
# Convert the GDP value in Million USD to Billion USD
df3[['GDP (Million USD)']]=df3[['GDP (Million USD)']]/1000
# Use numpy.round() method to round the value to 2 decimal places.
df3['GDP (Million USD)']=np.round(df3['GDP (Million USD)'],2)
# Rename the column header from 'GDP (Million USD)' to 'GDP (Billion USD)'
df3.rename(columns = {'GDP (Million USD)' : 'GDP (Billion USD)'})
df3
```

```python
# Load the DataFrame to the CSV file named "Largest_economies.csv"
fileName="./Largest_economies.csv"
df3.to_csv(fileName)
print("file saved")
```
